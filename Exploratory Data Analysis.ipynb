#### Exploratory Data Analysis

```python
# shape of the dataset
dataset.shape
```

```python
# Detecting what percent of missing values in Dataset columns
feature_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]

for feature in feature_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(),4),'% Missing Values')
```

```python
# Since they are many missing values, we need to find the relationship between missing values and Sales Price.
# Using Bar chart 
for feature in feature_with_na:
    data = dataset.copy()

    data[feature] = np.where(data[feature].isnull(),1,0)

    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
print("Id Of Houses {}".format(len(dataset.Id)))
```

```python
# Detecting the Number of Numerical columns are Present in dataset
Numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']

print("Numerical Values are: ", len(Numerical_features))


dataset[Numerical_features].head()
```

```python
# Detecting How many Year columns Are present in dataset
year_feature = [feature for feature in Numerical_features if 'Yr'in feature or 'Year' in feature]
year_feature
```

```python
# let's explore the content of these year variables
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
# Creating Line Chart for YearSold and SalePrice
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median of House Sold')
plt.title('House Price vs Year Sold')
plt.show()
```

```python
year_feature
```

```python
for feature in year_feature:
    if feature != 'YrSold':
        data = dataset.copy()

        data[feature] = data['YrSold']-data[feature]

        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
## Numerical variables are usually of 2 type
## 1. Continous variable and Discrete Variables
discrete_features = [feature for feature in Numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+["Id"]]
print("Discrete Columns Counts: {}".format(len(discrete_features)))
```

```python
discrete_features
```

```python
dataset[discrete_features].head()
```

```python
## Lets Find the realtionship between them and Sale PRice
for feature in discrete_features:
    data = dataset.copy()

    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
Continuous_feature = [feature for feature in Numerical_features if feature not in discrete_features+year_feature+["Id"]]
print("Continuous Features: {}".format(len(Continuous_feature)))
```

```python
for feature in Continuous_feature:
    data = dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Counts")
    plt.title(feature)
    plt.show()
```

```python
for feature in Continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
        
```

```python
for feature in Continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
